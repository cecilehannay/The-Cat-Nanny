# ❓ FAQ
___
## Questions Your Human Will Ask

**Will you meet me first?**  
Yes. A meet-and-greet is required.  
Your human can talk. You can observe. Approval is mutual.

**Will you follow my routine exactly?**  
Yes. Your routine is not “a suggestion.”  
Food, litter, timing, access rules—followed as written.

**What if I hide?**  
That is a valid choice and it will be respected.
I am very good at confirming your continued existence without making it weird.

**What if I live with other cats?**  
That happens. I can manage group dynamics, separate feeding, and respectful boundaries. If one of you is a food thief, I will notice.

**What if there’s an emergency?**  
Your human will provide emergency contacts and vet info.  
If something looks wrong, your human will be notified promptly and calmly.

**Will you give me my medicine?**  
Possibly. We will discuss this with your human. If you have strong opinions about pills, we will plan accordingly.

**Do you do overnight stays?**  
Currently: no. I specialize in in-home visits that preserve routine and dignity.  
If your situation is special, your human may inquire.

**Where I am willing to travel?**  
I currently serve: Louisville, Lafayette, South Boulder, East Boulder
If your residence is nearby and uncertain, your human may inquire.

**Do you care for other animals?**  
This site is for cats. If a small, quiet roommate exists (fish, etc.), we can discuss basic support.

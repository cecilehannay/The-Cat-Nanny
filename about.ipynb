## THE CAT NANNY

I have never met a cat I didn’t like me. Challenge me. Many cats accept me immediately. Others prefer a longer evaluation period. I respect both approaches.

I am calm, reliable, and observant. I understand that, like all cats, you have high standards, I follow instructions exactly as given by your human and do not improvise.

When I am not caring for you, I work as an atmospheric scientist at NCAR, a role that requires patience, precision, and careful attention to complex systems. These skills translate well.

No surprises. No improvisation.

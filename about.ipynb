# 😺 ABOUT US
____
## THE CAT NANNY

I have never met a cat I didn’t like. Challenge me. 

Many cats accept me immediately. Others prefer a longer evaluation period. I respect both approaches.

I am reliable, and attentive. I understand that, like all cats, you have high standards, I follow instructions exactly as given by your human and do not improvise.

When I am not caring for you, I work as an atmospheric scientist at NCAR, a role that requires patience, precision, and careful attention to complex systems. These skills translate well to cat care.

____

## THE APPROVED BACK UP

Genes do not lie. The son of the Cat Nanny is also excellent with cats. He has been raised in a household where cats are listened to and respected.

He is currently a senior honors undergraduate student in biochemistry, which means he is detail-oriented, reliable, and comfortable following precise instructions. 

On rare occasions, if I am fully booked, he may be available to substitute — **only with your approval and your human’s approval**.

No surprises. No improvisation.
____

## THE REAL BOSS

She is the queen of the house and our business supervisor. She doesn't participate in visits but she approves the standards, audits the staff, and decides what is acceptable.

Her expectations are clear:
- The food must be perfect
- The timing must be precise
- The staff must remain attentive

Approval is not automatic. Judgment is ongoing.

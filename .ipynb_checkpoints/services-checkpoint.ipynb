# 🐈 SERVICES & RATES
## What Will Happen During a Visit
Visits are adapted to your personality, not a checklist.
____

### 30-Minute Visit — $20
*Suitable for cats who prefer efficiency*
I will:
- Make food appear
- Refresh water
- Clean the litter box (again)
- Offer play, petting, or quiet coexistence
____
### 60-Minute Visit — $30
*Suitable for cats who value company*
I will:
- Do all of the above
- Stay longer, if you permit
- Provide additional playtime, petting or calm company
____
### Free Updates - $0
*To help your human relax*
Your human worries and misses you. This is expected.  
For no extra fee, I will: 
- Report to your human that all is well
- I will send photos and short videos so they know you remain in control of the situation
This service is for them. You may ignore it.

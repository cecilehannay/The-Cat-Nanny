# ❓ FAQ
## Questions Your Human Will Ask

**Will you meet me first?**  
Yes. Your approval is required.

**Will you follow my routine exactly?**  
Yes. Routines are not suggestions.

**What if I hide?**  
That is acceptable. I am very good at confirming your continued existence.

**Will you take care of my siblings?**  
Yes, I can do several cats and make sure that not one of them eat everything. 

**Will you give me my medicine?**  
Possibly. We will discuss this with your human.

**Are you doing dogs too?**  
Sorry. I am not a dog person (except for huskies)
